```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear el grafo
G = nx.DiGraph()

# Nodos para inputs multimodales
inputs = [
    "Texto (EN)", "Texto (ZH)", "Imágenes",
    "Audio", "Código", "Matemáticas"
]

# Nodo central del Semantic Hub
semantic_hub = "Semantic Hub"

# Nodos de salida
outputs = [
    "Salida: Subtítulos (Texto)",
    "Salida: Representación textual",
    "Salida: Expresión en palabras"
]

# Agregar nodos y conexiones al grafo
for input_modality in inputs:
    G.add_edge(input_modality, semantic_hub)

for output in outputs:
    G.add_edge(semantic_hub, output)

# Usar un layout más estructurado (shell layout)
pos = nx.shell_layout(G)

# Configuración de colores para nodos
input_colors = "deepskyblue"
hub_color = "darkorange"
output_colors = "mediumseagreen"

# Dibujar el grafo
plt.figure(figsize=(14, 10))
nx.draw_networkx_nodes(G, pos, nodelist=inputs, node_color=input_colors, node_size=2500, label="Inputs")
nx.draw_networkx_nodes(G, pos, nodelist=[semantic_hub], node_color=hub_color, node_size=4000, label="Semantic Hub")
nx.draw_networkx_nodes(G, pos, nodelist=outputs, node_color=output_colors, node_size=2500, label="Outputs")

# Dibujar las conexiones con un diseño más claro
nx.draw_networkx_edges(G, pos, edge_color="gray", arrowsize=20, arrowstyle="-|>", width=2)

# Etiquetas personalizadas con posiciones ajustadas
labels = {n: n for n in G.nodes()}
label_positions = {k: (v[0], v[1] + 0.1) for k, v in pos.items()}  # Elevar las etiquetas un poco
nx.draw_networkx_labels(G, label_positions, labels=labels, font_size=12, font_color="black", font_weight="bold")

# Anotaciones y estilo adicional
plt.title("Mapa Semántico: Convergencia Multimodal en el Semantic Hub", fontsize=16, pad=20)
plt.legend(scatterpoints=1, loc='upper left', fontsize=12)
plt.axis('off')
plt.show()
```
